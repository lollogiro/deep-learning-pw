# Deep Learning Project Work - Benchmarking efficient modern CNNs for micro-meteorological image classification

This project focuses on the automated classification of micro-weather conditions using images sourced from wildlife camera traps. The primary goal is to modernize the benchmarks established in recent ecological literature by evaluating the performance of the latest efficient Convolutional Neural Networks (CNNs) against legacy architectures.

## Dataset
The study utilizes the dataset associated with the paper ["Micro-weather classifications from wildlife cameras"](https://onlinelibrary.wiley.com/doi/10.1111/gcb.17078) (available on [Zenodo](https://zenodo.org/records/10137731)). It consists of data collected from 49 wildlife cameras located in the Maloti-Drakensberg (South Africa) and the Swiss Alps. The task involves classifying four distinct weather conditions: `overcast`, `sunshine`, `hail`, and `snow`. The dataset presents real-world challenges such as variable illumination, diverse backgrounds, and class imbalance.

## Modern baselines (MobileNetV3 & V4)
The original study established baselines using MobileNetV1 and MobileNetV2. This project benchmarks the latest iterations of the MobileNet family.

The specific models evaluated are:

**MobileNetV3 small**: [timm/mobilenetv3_small_100.lamb_in1k](https://huggingface.co/timm/mobilenetv3_small_100.lamb_in1k)

**MobileNetV4 small**: [timm/mobilenetv4_conv_small.e2400_r224_in1k](https://huggingface.co/timm/mobilenetv4_conv_small.e2400_r224_in1k)

These models were selected to compare performance with the similar small models trained in the paper.

## Analyzing dataset structure

Attached to the dataset there are CSV files with metadata for images and cameras. The main file for images is `image_data.csv`, with these columns, not all of which are needed/useful for model training:

- `DatasetOutFileName`: image file name.
- `label_sample`: sampling strategy used to select the image (not needed for model training).
- `Region`: camera region (e.g. Maloti-Drakensberg `ZA` or Swiss Alps `CH`).
- `cam`: unique camera ID that captured the image.
- `DateTimeOriginal`: original date/time from image metadata (raw format, may contain colons).
- `DateTime`: reformatted date/time (standardized representation).
- `DOY`: day of year ($1–366$).
- `Hour`: hour of capture ($0–23$).
- `Minute`: minute of capture ($0–59$).
- `Temperature`: ambient temperature recorded by the sensor at capture time.
- `Flash`: whether flash fired ($1$) or not ($0$).
- `ExposureTime`: shutter exposure time in seconds (decimal).
- `weather`: numeric weather class label (target for the model, e.g. 0–3).
- `weather_desc`: textual weather class (e.g. `background`, `sunshine`, `hail`, `snow`).
- `cis_fold`: fold index ($0–4$) for **Cis fivefold cross‑validation**, a within‑domain, image‑level random split where images from all cameras are randomly assigned to folds. Folds are mutually exclusive, stratified by the four study sites, and each fold is iteratively used as validation while the other four form the training set.
- `trans_fold`: fold index ($0–4$) for **Trans fivefold cross‑validation**, a camera‑level split where entire cameras (all their images) are assigned to folds so that no camera appears in both training and validation. Folds are mutually exclusive and stratified by the four study sites.
- `holdout`: boolean indicator (`TRUE` if reserved exclusively for the final test set).
- `cis_prod_fold`: alternative Cis fold variant (also a fivefold, site‑stratified Cis split) used for production/alternate experiment splits or final model selection; similar semantics to `cis_fold` but intended for deployment/production evaluation.

## Utils functions for splits creation

```python
def make_splits(df: pd.DataFrame, fold_col: str, test_fold_index: int, val_fold_index: int) -> Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    '''
    Function to split DataFrame into train/val/test sets according to specified fold column and index.
    Inputs:
    - df: input DataFrame
    - fold_col: name of the column containing fold assignments
    - fold_index: index of the fold to use as validation set
    Outputs:
    - train_df: training set DataFrame
    - val_df: validation set DataFrame
    - test_df: test set DataFrame
    '''
    assert fold_col in df.columns, f"Column '{fold_col}' not found in DataFrame."
    assert 1 <= test_fold_index <= 6, "test_fold_index must be between 1 and 6."
    assert 1 <= val_fold_index <= 6, "val_fold_index must be between 1 and 6."
    assert test_fold_index != val_fold_index, "test_fold_index and val_fold_index must be different."

    df_cp = df.copy()

    # fold_index as test set
    test_mask = df_cp[fold_col] == test_fold_index
    test_df = df_cp[test_mask].reset_index(drop=True)
    
    # fold_index - 1 as validation set (6 if fold_index is 1)
    val_fold_index = val_fold_index
    val_mask = df_cp[fold_col] == val_fold_index
    val_df = df_cp[val_mask].reset_index(drop=True)
    
    train_df = df_cp[~test_mask & ~val_mask].reset_index(drop=True)

    return train_df, val_df, test_df

# NOTE: not used to be consistent with the paper approach
def make_cis_splits(df: pd.DataFrame, test_index: int = 0, val_index: int = 1):
    return make_splits(df, 'cis_fold', test_fold_index=test_index, val_fold_index=val_index)

def make_trans_splits(df: pd.DataFrame, test_index: int = 0, val_index: int = 1):
    return make_splits(df, 'trans_fold', test_fold_index=test_index, val_fold_index=val_index)

# NOTE: not used to be consistent with the paper approach
def make_cis_prod_splits(df: pd.DataFrame, test_index: int = 0, val_index: int = 1):
    return make_splits(df, 'cis_prod_fold', test_fold_index=test_index, val_fold_index=val_index)
```

## Dataset Analysis

**1. Class and site imbalance**
I observed a significant imbalance in the dataset regarding both class frequencies and site contributions. Common classes like `sunshine` and `overcast` are heavily over-represented compared to rare events like `hail` and `snow`. Similarly, the image count is not uniformly distributed across the different camera sites.

**2. Split strategy and data leakage**
To ensure a fair comparison with the original paper, I adopted the `trans` split strategy provided by the authors. Data was split based on distinct camera locations (sites). This approach ensures no data leakage, as the model never encounters the same background context in both training and testing. This is essential for evaluating the model's ability to generalize to unseen locations.

**3. Temporal distribution**
* **Hourly:** The dataset is relatively well-balanced across hours, covering various lighting conditions.
* **Seasonal:** The day of year (DOY) distribution is non-uniform, showing specific gaps (e.g., around day 150 and 300) and peaks (e.g., around day 100 and year-end). These irregularities likely reflect the camera deployment schedule, which appears biased towards winter/spring months to capture a sufficient number of rare seasonal classes, such as `snow` and `hail`.

## Load dataset

Because the dataset identifies images using only their file names (without paths), and images are organized in subdirectories by class ID, a function is needed to resolve the full image paths based on the `DatasetOutFileName` and the directory structure.

```python
load_local_images = False

if load_local_images:

    file_map = defaultdict(list)
    for root, _, files in os.walk(IMAGES_DIR):
        for fname in files:
            if fname.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                file_map[fname].append(os.path.abspath(os.path.join(root, fname)))

    def resolve_filename(fname):
        hits = file_map.get(fname)
        return hits[0] if hits else None

    df_metadata_images['DatasetAbsPath'] = df_metadata_images['DatasetOutFileName'].map(resolve_filename)

    TARGET_SIZE = 518

    def process_image_png(args):
        src_path, rel_path = args
        if not src_path: return None
        
        dst_path = os.path.join(IMAGES_DIR_LOCAL, os.path.splitext(rel_path)[0] + ".png")
        
        if os.path.exists(dst_path):
            return dst_path
            
        try:
            os.makedirs(os.path.dirname(dst_path), exist_ok=True)
            with Image.open(src_path) as img:
                img = img.convert('RGB')
                img_resized = img.resize((TARGET_SIZE, TARGET_SIZE), Image.Resampling.LANCZOS)
                img_resized.save(dst_path, "PNG", compress_level=3)
            return dst_path
        except Exception as e:
            return None

    tasks = []
    for _, row in df_metadata_images.iterrows():
        tasks.append((row['DatasetAbsPath'], row['DatasetOutFileName']))

    with ProcessPoolExecutor(max_workers=8) as executor:
        local_paths = list(tqdm(executor.map(process_image_png, tasks), total=len(tasks)))

    df_metadata_images['DatasetLocalPath'] = local_paths
    df_metadata_images.to_csv(IMAGES_METADATA_CSV_LOCAL, index=False)
else:
    df_metadata_images = pd.read_csv(IMAGES_METADATA_CSV_LOCAL)
```

```python
class WeatherDataset(Dataset):
    def __init__(self, metadata, transform=None, local=True):
        self.metadata = metadata.reset_index(drop=True)
        self.image_paths = self.metadata['DatasetLocalPath'].tolist() if local else self.metadata['DatasetAbsPath'].tolist()
        self.labels = self.metadata['weather'].tolist()
        self.transform = transform
        
    def __len__(self):
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        max_retries = 5
        
        for attempt in range(max_retries):
            try:
                image = Image.open(image_path).convert("RGB")
                
                if self.transform:
                    image = self.transform(image)
                
                label = self.labels[idx]
                return image, label
                
            except Exception as e:
                if attempt == 0:
                    print(f"Error loading {image_path}: {e}")
                # try a different random sample
                idx = np.random.randint(0, len(self))
                image_path = self.image_paths[idx]
        
        raise RuntimeError(f"Failed to load any valid image after {max_retries} attempts")
```

## Utils functions

Utils function for both training and further evaluation

```python
def train_one_epoch(model, dataloader, criterion, optimizer, scaler, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    pbar = tqdm(dataloader, desc='train', leave=False)
    for images, labels in pbar:
        images = images.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        with torch.amp.autocast(enabled=(device.type=='cuda'), device_type=device.type):
            logits = model(images)
            loss = criterion(logits, labels)
        
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item() * images.size(0)
        preds = torch.argmax(logits, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
        pbar.set_postfix({'loss': running_loss/total, 'acc': correct/total})
    return running_loss/total, correct/total

def eval_one_epoch(model, dataloader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        pbar = tqdm(dataloader, desc='eval', leave=False)
        for images, labels in pbar:
            images = images.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            logits = model(images)
            loss = criterion(logits, labels)
            
            running_loss += loss.item() * images.size(0)
            preds = torch.argmax(logits, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            pbar.set_postfix({'loss': running_loss/total, 'acc': correct/total})
    return running_loss/total, correct/total

def save_checkpoint(model, optimizer, epoch, path):
    os.makedirs(os.path.dirname(path), exist_ok=True)
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
    }, path)

def load_best_model_eval(model_card, num_classes, checkpoint_path, device):
    model = timm.create_model(model_card, pretrained=False, num_classes=num_classes)
    
    checkpoint = torch.load(checkpoint_path, map_location=device)
    
    model.load_state_dict(checkpoint['model_state_dict'])
    
    model.to(device)
    model.eval()
    return model

def calculate_consensus_accuracy(ensemble_logits_stack, targets, min_agreement=5):
    individual_preds = torch.argmax(ensemble_logits_stack, dim=2).T
    
    mode_values, _ = torch.mode(individual_preds, dim=1)
    
    counts = (individual_preds == mode_values.unsqueeze(1)).sum(dim=1)
    
    sure_mask = counts >= min_agreement
    
    consensus_preds = mode_values
    
    total_samples = len(targets)
    kept_samples = sure_mask.sum().item()
    discarded_samples = total_samples - kept_samples
    
    if kept_samples == 0:
        print(f"Total: {total_samples} | Kept: 0 (0.0%)")
        return 0.0, 0.0
    
    correct_sure_preds = (consensus_preds[sure_mask] == targets[sure_mask]).sum().item()
    consensus_acc = correct_sure_preds / kept_samples
    coverage = kept_samples / total_samples
    
    print(f"Total: {total_samples}")
    print(f"Kept: {kept_samples} ({coverage*100:.1f}%)")
    print(f"Discarded: {discarded_samples}")
    print(f"Accuracy: {consensus_acc:.4f}")
    
    return consensus_acc, coverage

def plot_cm(cm, title, class_names, results_folder, filename):
    plt.figure(figsize=(10, 8))
    
    cm_norm = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-9)
    
    sns.heatmap(cm_norm, annot=True, fmt='.2%', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    
    plt.title(title, fontsize=15)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(os.path.join(results_folder, filename))
    plt.show()

def calculate_metrics(preds, targets):
    acc = accuracy_score(targets, preds)
    f1 = f1_score(targets, preds, average='macro')
    return acc * 100, f1 * 100

def analyze_results(results_folder):
    fold_stats = []

    total_cm_ensemble = None
    total_cm_consensus = None
    
    for fold_idx in range(1, N_FOLDS + 1):
        file_path = os.path.join(results_folder, f'fold_{fold_idx}_ensemble_data.pt')
        if not os.path.exists(file_path):
            print(f"WARNING: File not found {file_path}")
            continue

        data = torch.load(file_path, map_location='cpu')
        
        ensemble_logits_stack = data['logits'] # (5, N, C)
        targets = data['targets'].numpy()
        
        num_classes = ensemble_logits_stack.shape[2]
        if total_cm_ensemble is None:
            total_cm_ensemble = np.zeros((num_classes, num_classes), dtype=int)
            total_cm_consensus = np.zeros((num_classes, num_classes), dtype=int)
            
            global CLASS_NAMES
            if CLASS_NAMES is None or len(CLASS_NAMES) != num_classes:
                CLASS_NAMES = [str(i) for i in range(num_classes)]

        member_accs, member_f1s = [], []
        for i in range(ensemble_logits_stack.shape[0]):
            single_preds = torch.argmax(ensemble_logits_stack[i], dim=1).numpy()
            acc, f1 = calculate_metrics(single_preds, targets)
            member_accs.append(acc)
            member_f1s.append(f1)
        
        avg_logits = torch.mean(ensemble_logits_stack, dim=0)
        ens_preds = torch.argmax(avg_logits, dim=1).numpy()
        ens_acc, ens_f1 = calculate_metrics(ens_preds, targets)
        
        cm_fold_ens = confusion_matrix(targets, ens_preds, labels=range(num_classes))
        total_cm_ensemble += cm_fold_ens

        all_preds = torch.argmax(ensemble_logits_stack, dim=2)
        consensus_mask = (all_preds == all_preds[0]).all(dim=0).numpy() # Unanimità
        
        kept_samples = consensus_mask.sum()
        total_samples = len(targets)
        coverage = (kept_samples / total_samples) * 100
        
        if kept_samples > 0:
            cons_preds = all_preds[0][consensus_mask].numpy()
            cons_targets = targets[consensus_mask]
            cons_acc = accuracy_score(cons_targets, cons_preds) * 100
            
            cm_fold_cons = confusion_matrix(cons_targets, cons_preds, labels=range(num_classes))
            total_cm_consensus += cm_fold_cons
        else:
            cons_acc = 0.0

        fold_stats.append({
            'fold': fold_idx,
            'member_acc': np.mean(member_accs), 
            'member_f1': np.mean(member_f1s),
            'ens_acc': ens_acc, 
            'ens_f1': ens_f1,
            'cons_acc': cons_acc, 
            'coverage': coverage
        })

    df = pd.DataFrame(fold_stats)
    means, stds = df.mean(), df.std()

    print("-" * 100)
    print(f"{'Method':<20} | {'Macro-F1 (%)':<15} | {'Accuracy (%)':<15} | {'Consensus acc (%)':<18} | {'Coverage (%)':<15}")
    print("-" * 100)
    print(f"{'Ensemble':<20} | {means['ens_f1']:.2f} ({stds['ens_f1']:.2f})   | {means['ens_acc']:.2f} ({stds['ens_acc']:.2f})   | {means['cons_acc']:.2f} ({stds['cons_acc']:.2f})      | {means['coverage']:.2f} ({stds['coverage']:.2f})")
    print(f"{'Member (Avg)':<20} | {means['member_f1']:.2f} ({stds['member_f1']:.2f})   | {means['member_acc']:.2f} ({stds['member_acc']:.2f})   | {'-':<18} | {'-':<15}")
    print("-" * 100)
    
    plot_cm(total_cm_ensemble, 
            f'Standard ensemble\n(Accuracy: {means["ens_acc"]:.1f}%)', 
            CLASS_NAMES, results_folder, 'confusion_matrix_ensemble.png')

    plot_cm(total_cm_consensus, 
            f'Consensus ensemble\n(Accuracy: {means["cons_acc"]:.1f}% - Coverage: {means["coverage"]:.1f}%)', 
            CLASS_NAMES, results_folder, 'confusion_matrix_consensus.png')
```

## Training and evaluation of MobileNet models

To train these new versions of MobileNet, I introduced several modifications compared to the original paper to optimize convergence and robustness:

* **Hyperparameters:** unlike the original study, values for `learning_rate`, `weight_decay`, `dropout_rate`, and `batch_size` were determined empirically. These values were chose to ensure training convergence and maximize performance on the validation split.

* **Early Stopping:** `patience` parameter was significantly reduced from the original 50 epochs to **5**. This stricter stopping criterion helps in preventing overfitting and ensures that the model training stops as soon as the validation performance plateaus.

* **Augmentations:** the original paper limited augmentations to random horizontal/vertical flips and 45 degree rotations. I extended this pipeline to improve generalization:
    - `RandomCrop`: forcing the model to focus on local features rather than just the global context.
    - `ColorJitter`: introducing small variations in image illumination and saturation. Since weather classes are often strongly correlated with global color statistics (e.g., sunny images tend to be brighter/bluer), this prevents the model from relying solely on color histograms.
    - `RandomErasing`: randomly masking regions of the image. Simulating partial occlusions, I tried to prevent overfitting to specific visual patterns, may due to camera imbalances.
    - `Normalization`: standardizing input statistics to match the distribution used during the model's pre-training on ImageNet.

Also *gradient scaling* and *autocast* were employed to leverage mixed precision training, significantly reducing memory usage and accelerating training time, even if trained models are very small. Gradient scaling also ensured that small gradient values are preserved during backpropagation operations.

To maximize reproducibility and ensure a fair comparison, I assigned a unique deterministic seed to each training fold. I matched these seeds across both MobileNet models to control for stochastic variations in data shuffling and initialization.

### Common variables between training processes

```python
NUM_CLASSES = len(CLASS_NAMES)

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

LR_WARMUP = 1e-3
WARMUP_EPOCHS = 5
FINETUNE_EPOCHS = 30
PATIENCE = 5

MNV3_CARD = 'timm/mobilenetv3_small_100.lamb_in1k'

MNV4_CARD = 'timm/mobilenetv4_conv_small.e2400_r224_in1k'

N_FOLDS = 6
N_INNER_MODELS = 5

# the same across the two models
DATA_CONFIG = timm.data.resolve_data_config({}, model=MNV3_CARD)

IMG_SIZE = DATA_CONFIG['input_size'][1]

TRANSFORMS_TRAIN = T.Compose([
    T.Resize(int(IMG_SIZE / DATA_CONFIG['crop_pct']), interpolation=T.InterpolationMode.BICUBIC),
    T.RandomCrop(IMG_SIZE),
    T.RandomHorizontalFlip(), # as in the original paper
    T.RandomVerticalFlip(), # as in the original paper
    T.RandomRotation(degrees=45, interpolation=T.InterpolationMode.BICUBIC), # as in the original paper
    T.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    T.ToTensor(),
    T.Normalize(mean=DATA_CONFIG['mean'], std=DATA_CONFIG['std']),
    T.RandomErasing(p=0.5, scale=(0.02, 0.33), ratio=(0.3, 3.3), value=0)
])

TRANSFORMS_VAL = T.Compose([
    T.Resize(IMG_SIZE, interpolation=T.InterpolationMode.BICUBIC),
    T.CenterCrop(IMG_SIZE),
    T.ToTensor(),
    T.Normalize(mean=DATA_CONFIG['mean'], std=DATA_CONFIG['std'])
])
```

### MobileNetV3 - Nested Cross-Validation

```python
batch_size = 32
lr_finetune = 1e-4
weight_decay = 0.03
dropout_rate = 0.1

MNV3_FOLDER_RES = f'mobilenetv3_results/'

results_mobilenetv3 = []

for test_fold_idx in range(1, N_FOLDS + 1):
    print(f'\t\t\tOUTER FOLD {test_fold_idx}\n')
    
    available_folds = [f for f in range(1, N_FOLDS + 1) if f != test_fold_idx]
    
    fold_ensemble_logits = [] 
    test_targets = []
    
    for i, val_fold_idx in enumerate(available_folds):
        print(f'\tTraining inner model {i+1}/{N_INNER_MODELS} (test: {test_fold_idx}, val: {val_fold_idx})')

        current_seed = SEED + test_fold_idx + i
        fix_random(current_seed)
    
        g = torch.Generator()
        g.manual_seed(current_seed)
        
        train_df, val_df, test_df = make_trans_splits(
            df_metadata_images, 
            test_index=test_fold_idx, 
            val_index=val_fold_idx
        )
        
        train_ds = WeatherDataset(train_df, transform=TRANSFORMS_TRAIN)
        val_ds = WeatherDataset(val_df, transform=TRANSFORMS_VAL)
        test_ds = WeatherDataset(test_df, transform=TRANSFORMS_VAL)

        train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, 
                                  num_workers=4, pin_memory=True, persistent_workers=True,
                                 generator=g, worker_init_fn=seed_worker)
        val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False, 
                               num_workers=4, pin_memory=True, persistent_workers=True,
                               generator=g, worker_init_fn=seed_worker)
        test_loader = DataLoader(test_ds, batch_size=batch_size, shuffle=False, 
                                num_workers=4, pin_memory=True, persistent_workers=True,
                                generator=g, worker_init_fn=seed_worker)
        
        model = timm.create_model(
            MNV3_CARD, 
            pretrained=True, 
            num_classes=NUM_CLASSES, 
            drop_rate=dropout_rate,
            cache_dir="models-cache"
        )
        model.to(DEVICE)
        
        criterion = nn.CrossEntropyLoss()
        scaler = torch.amp.GradScaler(enabled=(DEVICE.type=='cuda'))

        # WARM-UP
        for name, param in model.named_parameters():
            if 'head' not in name: param.requires_grad = False

        optimizer = optim.AdamW(model.parameters(), lr=LR_WARMUP)
        
        for epoch in range(WARMUP_EPOCHS):
            train_one_epoch(model, train_loader, criterion, optimizer, scaler, DEVICE)
        
        eval_loss, val_acc = eval_one_epoch(model, val_loader, criterion, DEVICE)
            
        # FINE-TUNING
        for param in model.parameters(): param.requires_grad = True

        optimizer = optim.AdamW(model.parameters(), lr=lr_finetune, weight_decay=weight_decay)
        scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=FINETUNE_EPOCHS)

        best_val_acc = 0.0
        patience_it = 0
        model_save_path = os.path.join(MNV3_FOLDER_RES, f'test_{test_fold_idx}_val_{val_fold_idx}_best.pt')

        for epoch in range(FINETUNE_EPOCHS):
            train_one_epoch(model, train_loader, criterion, optimizer, scaler, DEVICE)
            val_loss, val_acc = eval_one_epoch(model, val_loader, criterion, DEVICE)
            
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                patience_it = 0
                save_checkpoint(model, optimizer, epoch, model_save_path)
            else:
                patience_it += 1
                
            scheduler.step()
            if patience_it >= PATIENCE: break

        model = load_best_model_eval(MNV3_CARD, NUM_CLASSES, model_save_path, DEVICE)
        
        all_logits = []
        all_targets = []
        model.eval()

        with torch.no_grad():
            for inputs, targets in test_loader:
                inputs = inputs.to(DEVICE)
                outputs = model(inputs)
                all_logits.append(outputs.cpu())
                all_targets.append(targets)
        
        fold_ensemble_logits.append(torch.cat(all_logits))
        
        if i == 0: # targets saved only once
            test_targets = torch.cat(all_targets)

        del model, optimizer, scaler, scheduler, criterion
        torch.cuda.empty_cache()

    ensemble_logits_stack = torch.stack(fold_ensemble_logits)
    avg_logits = torch.mean(ensemble_logits_stack, dim=0)
    
    preds = torch.argmax(avg_logits, dim=1)
    ensemble_acc = (preds == test_targets).float().mean().item()
    
    print(f'\t\tTest fold {test_fold_idx} | Ensemble accuracy: {ensemble_acc:.4f}\n')
    
    results_mobilenetv3.append({
        'test_fold': test_fold_idx, 
        'ensemble_acc': ensemble_acc,
        'logits_path': os.path.join(MNV3_FOLDER_RES, f'fold_{test_fold_idx}_ensemble_logits.pt')
    })
    
    torch.save({
        'logits': ensemble_logits_stack, 
        'avg_logits': avg_logits,
        'targets': test_targets
    }, os.path.join(MNV3_FOLDER_RES, f'fold_{test_fold_idx}_ensemble_data.pt'))

df_results_mobilenetv3 = pd.DataFrame(results_mobilenetv3)

print('\nAll folds ensemble results:')
print(df_results_mobilenetv3)
print('\nMean Accuracy')
print(df_results_mobilenetv3['ensemble_acc'].mean())

df_results_mobilenetv3.to_csv(os.path.join(MNV3_FOLDER_RES, 'ensemble_results.csv'))
```

### Consensus accuracy analysis for MobileNetV3


```python
min_agreement = 5

print(f"Minimum agreement required: {min_agreement}/5 models\n")

consensus_results_mnv3 = []

for fold_idx in range(1, 7):
    print(f"\tFOLD {fold_idx}\n")
    
    ensemble_data_path = os.path.join(MNV3_FOLDER_RES, f'fold_{fold_idx}_ensemble_data.pt')
    data = torch.load(ensemble_data_path)
    
    ensemble_logits_stack = data['logits']
    test_targets = data['targets']
    
    cons_acc, coverage = calculate_consensus_accuracy(ensemble_logits_stack, test_targets, min_agreement=min_agreement)
    
    consensus_results_mnv3.append({
        'fold': fold_idx,
        'consensus_accuracy': cons_acc,
        'coverage': coverage
    })
    print("\n")

df_consensus_mnv3 = pd.DataFrame(consensus_results_mnv3)
print(df_consensus_mnv3)
print(f"\nMean Consensus Accuracy: {df_consensus_mnv3['consensus_accuracy'].mean():.4f}")
print(f"Mean Coverage: {df_consensus_mnv3['coverage'].mean():.4f}")
```

## Discussion

In order to maintain a valid comparison with results in the original paper, I strictly adhered to the same evaluation methodology, specifically utilizing **Cross-Validation Aggregation** (often referred to as *crogging*).

Instead of relying on a single model, this procedure involves constructing ensembles from the sets of five "member models" generated during the inner cross-validation loop. In this setup, every observation in the training set contributes to the training of four member models and serves as the early-stopping validation set for the fifth. This technique is known to improve both generalization performance and model stability by averaging out stochastic variations. For each of the outer "Trans" validation folds, the five corresponding member models were aggregated by performing an unweighted average of their output softmax probabilities.

By observing the obtained results, we can conclude that the performance varies significantly between the two modern architectures tested.

### MobileNetV3
This model demonstrated robust performance, achieving a *mean ensemble accuracy* of **94.15%** across all folds, effectively outperforming the baselines established in the original study. This performance was further enhanced by applying the **consensus method**, where a prediction is only made if all 5 member models agree on the label. This strict criterion raised the accuracy to **97.71%**, while discarding **16.6%** of the samples (mean coverage of 83.40%). This trade-off highlights the model's ability to identify ambiguous samples, ensuring high reliability for the retained predictions.

Analyzing the specific folds, `Fold 6` (often used as a holdout test set in the original configuration) proved to be the most challenging for MobileNetV3, resulting in the lowest ensemble accuracy of **92.72%**, closely followed by `Fold 2` with **92.73%**.

### MobileNetV4
Contrary to expectations, the newer **MobileNetV4** did not outperform its predecessor in this specific task, achieving a lower *mean ensemble accuracy* of **91.28%**. While the **consensus method** was able to recover a high accuracy of **96.97%**, it came at a significant cost: the model had to discard nearly **28%** of the samples (mean coverage of 72.09%) to achieve this reliability, a performance notably inferior to that of MobileNetV3.

Regarding the training strategy, I observed that MobileNetV4 achieved superior convergence without a "head warm-up" (linear probing) phase, allowing for a faster adaptation of the backbone features. However, even with this optimization, the results remained suboptimal. This outcome suggests that the hyperparameters chosen, while effective for V3, may not be suitable for this newer architecture. Consequently, a more extensive and structured optimization phase, such as a systematic grid-search, would likely be required to identify the ideal configuration and unlock the full potential of MobileNetV4 for this specific dataset.

### Class-wise Analysis
For MobileNetV3, the confusion matrix reveals that the most frequent error occurs in distinguishing `sunshine` from `overcast`, with **11.48%** of sunny images misclassified as overcast. Visual inspection of the dataset confirms that the difference between sunny and cloudy images is often subtle, relying on global color statistics that are visually very close. Similarly, MobileNetV4 also faced challenges with the `sunshine` class, suggesting this is a common difficulty across both architectures.

However, MobileNetV4 struggled significantly more with the under-represented classes compared to V3. While V3 maintained high accuracy for rare events, V4 frequently misclassified `hail` as `overcast` (17.22%) or `snow` (7.84%).
* *MobileNetV3* achieved a classification accuracy of **93.80%** for `snow` and **90.14%** for `hail`.
* *MobileNetV4* reached only **88.29%** for `snow` and **74.58%** for `hail`.

### Hyperparameter observations
Regarding the selected hyperparameters, we can observe that strong regularization was necessary for maximizing the performance of these lightweight models. `dropout` and a relatively high `weight decay` proved essential, likely counteracting the risk of overfitting inherent to a dataset containing many visually similar images. Additionally, a smaller `batch size` of **32** consistently yielded better generalization than larger batches; this suggests that the noise introduced by smaller batch gradient updates acted as an effective regularizer, helping the model to escape sharp minima. Indeed, with larger batch sizes like 64, I observed results merely comparable to the baseline in the paper. Finally, a similar learning rate value was found to be effective for both architectures.

## Conclusion
In conclusion, **MobileNetV3 small** offers the best trade-off between coverage and accuracy for this meteorological task. It provides higher classification confidence and superior detection of rare weather events (`hail` and `snow`) compared to the newer *MobileNetV4*, making it the preferable choice for automated environmental monitoring in this context, particularly for deployments on constrained hardware.

### Future Developments
To further validate and extend the findings of this study, several future research directions are proposed:

* **External dataset validation:** to verify the generalization capabilities of the models beyond the specific conditions of the current dataset, the evaluation should be extended to standard benchmarks such as the [CUHK Weather Dataset](https://www.cse.cuhk.edu.hk/leojia/projects/weatherclassify/index.htm). This would align the benchmarking process with the methodology adopted in the original paper, ensuring a more robust comparison.
* **Extended Cross-Validation:** while this experiment focused on the *Trans-location* strategy, future work could expand the analysis to include alternative validation splits (e.g., *Cis-location* or time-split folds) to assess model robustness under different temporal and spatial constraints.
* **Evaluation of MobileNetV5:** with the recent introduction (October 2025) of the **MobileNetV5** architecture, it will be crucial to benchmark it against V3 and V4. However, as the currently available versions (e.g., 300M parameters) are too large for a fair comparison in edge-computing scenarios, this step is deferred until lightweight "small" variants are released.
